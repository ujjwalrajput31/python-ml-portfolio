```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Dataset_CyberCrime_Sean.csv')
df.head(10)
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
plt.title('missing value haetmap')
plt.show()
```

```python
# Display the duplicate rows
duplicate_rows = df[df.duplicated()]
display(duplicate_rows)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
top_10=df.sort_values(by='Total',ascending=False).head(30)
plt.figure(figsize=(10, 6)) # Increase figure size
sns.barplot(x='Total',y='City',data=top_10,palette='Reds_r', hue='City', legend=False)
plt.title('Top 10 cities with most cyber crimes')
plt.xlabel('Total Cases')
plt.ylabel('City')
plt.show()
```

```python
# Group by City and sum the Total cases
city_total_crimes = df.groupby('City')['Total'].sum().reset_index()

# Sort by total crimes in descending order and select the top 10 cities
top_10_cities = city_total_crimes.sort_values(by='Total', ascending=False).head(10)

# Display the top 10 cities
display(top_10_cities)

# Generate a bar plot for the top 10 cities
plt.figure(figsize=(10, 6))
sns.barplot(x='Total', y='City', data=top_10_cities, palette='Reds_r', hue='City', legend=False)
plt.title('Top 10 Unique Cities with Most Cyber Crimes')
plt.xlabel('Total Cases')
plt.ylabel('City')
plt.show()
```
